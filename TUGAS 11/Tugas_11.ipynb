```python
import datetime
import os
import string
import random

# Template dictionary buku
buku_template = {
    'judul': 'judul',
    'penulis': 'penulis',
    'tahun_terbit': 0,
    'tanggal_pinjam': datetime.datetime(1111,1,11),
    'status': 'TERSEDIA'
}

data_buku = {}

while True:
    os.system('clear')
    print(f"{'SISTEM PERPUSTAKAAN':^50}")
    print(f"{'TAMBAH BUKU BARU':^50}")
    print("="*50)

    buku = dict.fromkeys(buku_template.keys())
    buku['judul'] = input("Judul Buku: ")
    buku['penulis'] = input("Penulis: ")
    buku['tahun_terbit'] = int(input("Tahun Terbit: "))

    TAHUN_PINJAM = int(input("Tahun Pinjam (YYYY): "))
    BULAN_PINJAM = int(input("Bulan Pinjam (1-12): "))
    TANGGAL_PINJAM = int(input("Tanggal Pinjam (1-31): "))
    buku['tanggal_pinjam'] = datetime.datetime(TAHUN_PINJAM, BULAN_PINJAM, TANGGAL_PINJAM)

    buku['status'] = "TERSEDIA"

    # Generate key unik
    KEY = ''.join(random.choice(string.ascii_uppercase) for i in range(6))
    data_buku.update({KEY: buku})

    # Tampilkan data
    print(f"\n{'KEY':<6} {'JUDUL':<20} {'PENULIS':<15} {'TAHUN':<6} {'TGL PINJAM':<12} {'STATUS':<10}")
    print("-"*70)

    for key in data_buku:
        JUDUL = data_buku[key]['judul']
        PENULIS = data_buku[key]['penulis']
        TAHUN = data_buku[key]['tahun_terbit']
        TGL_PINJAM = data_buku[key]['tanggal_pinjam'].strftime("%d-%m-%Y")
        STATUS = data_buku[key]['status']

        print(f"{key:<6} {JUDUL:<20} {PENULIS:<15} {TAHUN:<6} {TGL_PINJAM:<12} {STATUS:<10}")

    print("\n")
    tambah_lagi = input("Tambah buku lagi? (y/n): ")
    if tambah_lagi.lower() == 'n':
        break

print("Terima kasih telah menggunakan sistem perpustakaan!")
```
